```python
import torch
from torch import nn, optim 
import torch.nn.functional as F
from torchvision import datasets,transforms,models
from collections import OrderedDict
```

```python
data_directory = "dataset"
tr_train = transforms.Compose([transforms.RandomRotation(60),
                               transforms.RandomResizedCrop(225),
                               transforms.RandomHorizontalFlip(),
                               transforms.ToTensor(),
                               transforms.Normalize([0.485, 0.456, 0.406],
                                                   [0.229, 0.224, 0.225])])
tr_test = transforms.Compose([
                              transforms.Resize(250,250),
                              transforms.CenterCrop(224),
                              transforms.ToTensor(),
                              transforms.Normalize([0.485, 0.456, 0.406],
                                                 [0.229, 0.224, 0.225])])

train_data = datasets.ImageFolder(data_directory+'/train',transform = tr_train)
test_data = datasets.ImageFolder(data_directory+'/test',transform = tr_test)
trainloader = torch.utils.data.DataLoader(train_data,batch_size=32,shuffle=True)
testloader = torch.utils.data.DataLoader(test_data,batch_size=32)
```

```python
model = models.resnet101(pretrained = True)
model
```
